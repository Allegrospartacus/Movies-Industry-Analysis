# correlation matrix for all features

first we need to converts the categorical data into numeric representations.

a few ways to do that are :
1. Factorize Method (for quick analysis) and that which i will use
2. One-Hot Encoding (for deeper analysis)
3. Label Encoding (if you assume order matters)

after finishing the project i find out that .cat.codes could be better than Factorize Method in this case  

```python
df_factorized = df.apply(lambda x: x.factorize()[0] if x.dtype == 'object' else x)

correlation_matrix = df_factorized.corr()
correlation_matrix
```

```python
plt.figure(figsize=(15, 7.5))

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for all Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

just for simplesity Find strong and very strong positive and negative correlations

```python

# Strong positive correlations (0.5 <= correlation < 0.7)
strong_positive_correlations = correlation_matrix[(correlation_matrix >= 0.5) & (correlation_matrix < 0.7)]

# Very strong positive correlations (correlation >= 0.7)
very_strong_positive_correlations = correlation_matrix[correlation_matrix >= 0.7]

# Strong negative correlations (-0.7 < correlation <= -0.5)
strong_negative_correlations = correlation_matrix[(correlation_matrix <= -0.5) & (correlation_matrix > -0.7)]

# Very strong negative correlations (correlation <= -0.7)
very_strong_negative_correlations = correlation_matrix[correlation_matrix <= -0.7]


```

```python
#Remove self-correlations (where the correlation is 1.0 or -1.0 on the diagonal)
#Remove duplicate pairs (since correlation between A and B is the same as B and A)

# Helper function to process correlations
def process_correlations(correlation_df):
    correlation_df = correlation_df.where(~(correlation_df == 1) & ~(correlation_df == -1)).stack().reset_index()
    correlation_df.columns = ['Feature 1', 'Feature 2', 'Correlation']
    correlation_df = correlation_df[correlation_df['Feature 1'] != correlation_df['Feature 2']]
    correlation_df = correlation_df[~correlation_df[['Feature 1', 'Feature 2']].apply(frozenset, axis=1).duplicated()]
    return correlation_df


# Apply the Helper function to strong and very strong correlations
strong_positive_correlations = process_correlations(strong_positive_correlations)
very_strong_positive_correlations = process_correlations(very_strong_positive_correlations)
strong_negative_correlations = process_correlations(strong_negative_correlations)
very_strong_negative_correlations = process_correlations(very_strong_negative_correlations)

```

```python
# Output the correlations

print("Strong Positive Correlations (0.5 <= correlation < 0.7):")
print(strong_positive_correlations.sort_values(by='Correlation', ascending=False))

print("\nVery Strong Positive Correlations (correlation >= 0.7):")
print(very_strong_positive_correlations.sort_values(by='Correlation', ascending=False))

print("\nStrong Negative Correlations (-0.7 < correlation <= -0.5):")
print(strong_negative_correlations.sort_values(by='Correlation', ascending=False))

print("\nVery Strong Negative Correlations (correlation <= -0.7):")
print(very_strong_negative_correlations.sort_values(by='Correlation', ascending=False))
```
